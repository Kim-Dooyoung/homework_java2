```java
public record Pair<F extends Comparable<F>, S extends Comparable<S>>(F first, S second) 
        implements Comparable<Pair<F, S>> {

    // 제너릭 메소드 of를 사용하여 Pair를 생성
    public static <F extends Comparable<F>, S extends Comparable<S>> Pair<F, S> of(F first, S second) {
        return new Pair<>(first, second);
    }

    // 사전순으로 비교하도록 compareTo 메소드 구현
    @Override
    public int compareTo(Pair<F, S> other) {
        int cmp = this.first.compareTo(other.first);
        if (cmp == 0) {
            cmp = this.second.compareTo(other.second);
        }
        return cmp;
    }
}
```

```java
public class PairTest {
    public static void main(String[] args) {
        // Case 1: 왼쪽이 더 큰 경우
        Pair<Integer, String> pair1 = Pair.of(2, "apple");
        Pair<Integer, String> pair2 = Pair.of(1, "banana");
        System.out.println("Case 1 (left > right): " + pair1.compareTo(pair2)); // Expected: positive number

        // Case 2: 왼쪽이 더 작은 경우
        Pair<Integer, String> pair3 = Pair.of(1, "apple");
        Pair<Integer, String> pair4 = Pair.of(2, "banana");
        System.out.println("Case 2 (left < right): " + pair3.compareTo(pair4)); // Expected: negative number

        // Case 3: 같은 경우 (first는 같고 second가 다를 경우)
        Pair<Integer, String> pair5 = Pair.of(1, "apple");
        Pair<Integer, String> pair6 = Pair.of(1, "banana");
        System.out.println("Case 3 (same first, different second): " + pair5.compareTo(pair6)); // Expected: negative number

        // 추가 테스트: 같은 경우 (first와 second가 모두 같은 경우)
        Pair<Integer, String> pair7 = Pair.of(1, "apple");
        Pair<Integer, String> pair8 = Pair.of(1, "apple");
        System.out.println("Case 4 (same first and second): " + pair7.compareTo(pair8)); // Expected: 0
    }
}
```

```java
public class ModifiablePair<F extends Comparable<F>, S extends Comparable<S>> 
        implements Comparable<ModifiablePair<F, S>> {
    
    private Pair<F, S> pair; // Pair 인스턴스 필드

    // 생성자
    public ModifiablePair(F first, S second) {
        this.pair = Pair.of(first, second);
    }
    
    // getter는 pair에게 위임
    public F first() {
        return pair.first();
    }
    
    public S second() {
        return pair.second();
    }
    
    // setter를 정의
    public void setFirst(F first) {
        pair = Pair.of(first, second()); // 새로운 Pair를 생성
    }
    
    public void setSecond(S second) {
        pair = Pair.of(first(), second); // 새로운 Pair를 생성
    }

    // compareTo 메소드는 pair에게 위임
    @Override
    public int compareTo(ModifiablePair<F, S> other) {
        return this.pair.compareTo(other.pair);
    }

    @Override
    public String toString() { 
        return pair.toString(); // toString은 pair에게 위임
    }
}
```

```java
public class ModifiablePairTest {
    public static void main(String[] args) {
        // ModifiablePair 생성
        ModifiablePair<Integer, String> modPair = new ModifiablePair<>(1, "apple");

        // 초기 값 확인
        System.out.println("Initial values:");
        System.out.println("First: " + modPair.first()); // Expected: 1
        System.out.println("Second: " + modPair.second()); // Expected: "apple"

        // 값 변경
        modPair.setFirst(2);
        modPair.setSecond("banana");

        // 변경된 값 확인
        System.out.println("\nAfter setting new values:");
        System.out.println("First: " + modPair.first()); // Expected: 2
        System.out.println("Second: " + modPair.second()); // Expected: "banana"

        // compareTo 테스트
        System.out.println("\nComparing ModifiablePair instances:");

        ModifiablePair<Integer, String> pair1 = new ModifiablePair<>(2, "apple");
        ModifiablePair<Integer, String> pair2 = new ModifiablePair<>(1, "banana");
        ModifiablePair<Integer, String> pair3 = new ModifiablePair<>(2, "banana");
        ModifiablePair<Integer, String> pair4 = new ModifiablePair<>(2, "apple");

        // Case 1: 왼쪽이 더 큰 경우
        System.out.println("Case 1 (pair1 > pair2): " + pair1.compareTo(pair2)); // Expected: positive number

        // Case 2: 왼쪽이 더 작은 경우
        System.out.println("Case 2 (pair2 < pair1): " + pair2.compareTo(pair1)); // Expected: negative number

        // Case 3: 같은 경우 (first는 같고 second가 다를 경우)
        System.out.println("Case 3 (pair1 same first, different second): " + pair1.compareTo(pair3)); // Expected: negative number

        // 추가 테스트: 같은 경우 (first와 second가 모두 같은 경우)
        System.out.println("Case 4 (pair1 == pair4): " + pair1.compareTo(pair4)); // Expected: 0
    }
}
```
